resize(src, dsize[, dst[,fx[, fy[, interpolation]]]])
- src : required input image
- dsize : desired size of the output image 
- fx : scale factor along horizontal axis
- fy : scale factor along vertical axis 
- interpolation : it gives us the option of different methods of resizing the image 

```python
def rescaleFrame(frame, scale=0.75):
    width = int(frame.shape[1] * scale)
    height = int(frame.shape[0] * scale)
    dimension = (width,height)
    
    return cv.resize(frame, dimension, interpolation=cv.INTER_AREA)
```

# Image transformation 

translation , rotation, resizing, clipping and cropping 

```python
#Transalation 
def translation(img,x,y):
    transMat = np.float32([[1,0,x],[0,1,y]])
    dimensions = (img.shape[1], img.shape[0])
    
    return cv.warpAffine(img, transMat, dimensions)
# -x - left, -y up, x right, y down  
#  shift the image into left, right , up , down 
```

```python
translated = translation(img, 100, 100)
# cv.imshow("Translated",translated)
```

```python
#Rotation 
def rotate(img, angle, rotPoint = None):
    (height,width) = img.shape[:2]
    
    if rotPoint is None:
        rotPoint = (width//2, height//2)
    
    rotMat = cv.getRotationMatrix2D(rotPoint, angle, 1.0) #scale value
    dimensions = (width,height)
    
    return cv.warpAffine(img, rotMat, dimensions)
```

```python
rotated = rotate(img,-45)
# cv.imshow('Rotated',rotated)
```

Countours in OPENCV

```python
#using recoginition and 
img = cv.imread("cat_image.jpg")
cv.imshow("Cat",img)
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv.imshow('gray',gray)
cv.waitKey(0)
```

```python
canny = cv.Canny(img, 125, 175)
cv.imshow("Canny",canny)
cv.waitKey(0)
```

```python
#find countours based on 2 things canny and hierachies
contours, hierarchies = cv.findContours(canny, cv.RETR_LIST, cv.CHAIN_APPROX_NONE)
#return all coordinates of countours into value 
#return all hierarchies representations find the countours 
#cv.RETR_LIST / cv.RETRL_EXTERNAL /cv.RETR_TREE
print(f'{len(contours)} total of countours number found')
```

```python
#blur the image 
blur = cv.GaussianBlur(gray, (5,5), cv.BORDER_DEFAULT)
cv.imshow("Blur",blur)
cv.waitKey(0)

canny = cv.Canny(blur, 125, 175)
cv.imshow("Canny",canny)
cv.waitKey(0)

contours, hierarchies = cv.findContours(canny, cv.RETR_LIST, cv.CHAIN_APPROX_NONE)
print(f'{len(contours)} total of countours number found')
```

```python
ret, thresh = cv.threshold(gray, 175, 255,  cv.THRESH_BINARY)
# IF a particular pixel is below 125, if the density of the pixel is <125, set to 0 or blank
# if > 175, set to white or 255
contours, hierarchies = cv.findContours(thresh, cv.RETR_LIST, cv.CHAIN_APPROX_SIMPLE)
print(f'{len(contours)} total of countours number found')
```

#Bluring technique

```python
img1 = cv.imread("squat_barbell.jpg")
cv.imshow("img1",img1)
 #method 1 : averaging 
 #define a specific protion of an image : Divide into specific area of pixel / compute the  pixel intensities -> averaging the pixel intensities 
average = cv.blur(img1, (3,3))
cv.imshow("Average Blur",average)

 #method 2: Gaussian blur 
 #same with averaging , but each surroding has a particular weight -> Giving the true center the weight of pixels
gauss = cv.GaussianBlur(img1, (3,3), 0)
cv.imshow("Gaussian Blur",gauss)

 #method3 : Median 
 #same with the average , but finding the median -> Using the same intensities pixel of median
median = cv.medianBlur(img1, 3)
cv.imshow("Median Blủr",median)

 #method 4: Bilateral Bluring 
 #most effective -> applied bluring but retain the edges of the images
bilateral = cv.bilateralFilter(img1, 3, 15, 15)
cv.imshow("Bilateral",bilateral)

cv.waitKey(0)
```
